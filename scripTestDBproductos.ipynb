# [dbo].[sp_InsertarProducto]

```sql
/****** Object:  StoredProcedure [dbo].[sp_InsertarProducto]    Script Date: 17/7/2023 00:32:18 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_InsertarProducto]
(@TipoProductoId INT,
@Nombre NVARCHAR(150),
@Detalle NVARCHAR(200),
@Precio INT
)
AS
INSERT INTO Producto(TipoProductoId, Nombre, Detalle, Precio)
VALUES(@TipoProductoId,@Nombre,@Detalle,@Precio)
SELECT @@IDENTITY AS Id
RETURN
GO
```

# [dbo].[sp_ModificarProdcuto]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ModificarProdcuto]    Script Date: 17/7/2023 00:32:18 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_ModificarProdcuto]
(
@TipoProductoId INT,
@Nombre NVARCHAR(150),
@Detalle NVARCHAR(200),
@Precio INT,
@Id INT
)
AS
UPDATE Producto
SET TipoProductoId=@TipoProductoId,Nombre=@Nombre,Detalle=@Detalle,Precio=@Precio
WHERE IdProducto=@Id
GO
```
